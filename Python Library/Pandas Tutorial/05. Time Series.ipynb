```python
dataUrl = 'https://raw.githubusercontent.com/Datamanim/pandas/main/seoul_pm.csv'
df = pd.read_csv(dataUrl)
df.head()
```

```python
def change_date(x):
    import datetime
    date, time = x.split(':')
    
    if time == '24':
        time = '00:00:00'
        finalTime = pd.to_datetime(date + " " + time) + datetime.timedelta(days=1)
    else:
        time += ':00:00'
        finalTime = pd.to_datetime(date + " " + time)
    return finalTime

df['(년-월-일:시)'] = df['(년-월-일:시)'].apply(change_date)
df.head()
```

```python
dic = {
    0: 'Monday',
    1: 'Tuesday',
    2: 'Wedneday',
    3: 'Thursday',
    4: 'Friday',
    5: 'Saturday',
    6: 'Sunday'
}
df['dayName'] = df['(년-월-일:시)'].dt.weekday.map(lambda x: dic[x])
df['dayName']
```

```python
ans1 = df.groupby(['dayName', 'PM10등급'], as_index=False).size()
ans1
```

```python
ans2 = ans1.pivot(index='dayName', columns='PM10등급', values='size').fillna(0)
ans2
```
